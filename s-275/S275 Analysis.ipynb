```python
df_s275 = pd.concat([df_2005, df_2010, df_2015, df_2021], ignore_index=True)
df_s275_limited = pd.concat([df_2010, df_2015, df_2021], ignore_index=True)
```

```python
df_contractors = df_s275[df_s275['droot'].isin([63,64])]
df_teachers = df_s275[df_s275['droot'].isin([31,32,33,34,39,49,52,63,64])]
```

```python
# Define a function to calculate the proportions
def calculate_proportions(data):
    total_count = len(data)
    female_proportion = (data['sex'] == 'F').mean()
    masters_proportion = (data['hdeg'] == 'M').mean()
    white_proportion = (data['race'] == 'W').mean()
    return female_proportion, masters_proportion, white_proportion

# Create a PrettyTable
table = PrettyTable()
table.field_names = ["SchoolYear", "Subgroup", "Proportion Female", "Proportion Masters", "Proportion White"]

# Define the subgroups
subgroups = {
    "External Contracts": [63, 64],
    "All Teachers": [31, 32, 33, 34, 39, 49, 52, 63, 64]
}

# Calculate proportions for each SchoolYear and subgroup
for year in df_s275_limited['SchoolYear'].unique():
    for subgroup, droots in subgroups.items():
        filtered_data = df_s275_limited[(df_s275_limited['SchoolYear'] == year) & (df_s275_limited['droot'].isin(droots))]
        female_prop, masters_prop, white_prop = calculate_proportions(filtered_data)
        table.add_row([year, subgroup, f"{female_prop:.2%}", f"{masters_prop:.2%}", f"{white_prop:.2%}"])

# Print the table
print(table)
```

```python
df_contractors_year = df_contractors.groupby('SchoolYear').size().reset_index(name='Count')
df_teachers_year = df_teachers.groupby('SchoolYear').size().reset_index(name='Count')
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_contractors_year['SchoolYear'], df_contractors_year['Count'], marker='o', linestyle='-')
plt.title('Number of Observations of Contractor Teachers By School Year')
plt.xlabel('School Year')
plt.xticks(fontsize=8, rotation=45) 
plt.ylabel('Count')
plt.grid(True)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_teachers_year['SchoolYear'], df_teachers_year['Count'], marker='o', linestyle='-')
plt.title('Number of Observations of Teachers By School Year')
plt.xlabel('School Year')
plt.xticks(fontsize=8, rotation=45) 
plt.ylabel('Count')
plt.grid(True)
```

```python
df_unique_contractors_year = df_contractors.groupby('SchoolYear')['cert'].nunique()
df_unique_teachers_year = df_teachers.groupby('SchoolYear')['cert'].nunique()
```

```python
# Create line plots for total teachers
plt.figure(figsize=(8, 6))

# Plot for unique certifications of teachers
plt.plot(df_unique_teachers_year.index, df_unique_teachers_year.values, marker='o', linestyle='-', label='All Teachers')

plt.title('Year-over-Year Distribution of Teachers')
plt.xlabel('School Year')
plt.ylabel('Count of Unique Certifications')
plt.xticks(fontsize=8, rotation=45) 
plt.legend()
plt.grid(True)

# Show the plots
plt.show()
plt.savefig("all_teachers.png")
```

```python
# Create line plots for total teachers and total contractors
plt.figure(figsize=(8, 6))

# Plot for unique certifications of contractors
plt.plot(df_unique_contractors_year.index, df_unique_contractors_year.values, marker='o', linestyle='-', label='Contractor Teachers')

plt.title('Year-over-Year Distribution of Contractor Teachers')
plt.xlabel('School Year')
plt.ylabel('Count of Unique Certifications')
plt.xticks(fontsize=8, rotation=45) 
plt.legend()
plt.grid(True)

# Show the plots
plt.show()
plt.savefig("contractor_teachers.jpg")
```
